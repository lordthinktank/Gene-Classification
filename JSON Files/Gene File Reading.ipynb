```python
for i in range(2, 524):
    file_name = 'phages_gene_' + str(i) + '.json'
    with urllib.request.urlopen("http://phagesdb.org/api/genes/?page=" + str(i)) as url:
        with open(file_name, 'w') as outfile:
            json.dump(json.loads(url.read().decode()), outfile)
```

```python
df_list = []
for i in range(1, 524):
    with open('phages_gene_' + str(i) + '.json') as f:
        d = json.load(f)
    genes = json_normalize(d['results'])
    filter_genes = genes['GeneID'].str.contains('duplicate')
    genes_dropped = genes[~filter_genes]
    df_list.append(genes_dropped)
len(df_list)
```

```python
genes_concat = pd.concat(df_list, ignore_index=True)
```
